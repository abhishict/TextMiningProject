```r

library(arules)
library(arulesViz)

transactions <- read.transactions("/content/transactions_reviews.csv",
                                  format = "basket",
                                  sep = ",",
                                  rm.duplicates = TRUE)


summary(transactions)
itemFrequencyPlot(transactions, topN = 30, type = "absolute")

rules <- apriori(transactions,
                 parameter = list(supp = 0.01, conf = 0.25, minlen = 2))

inspect(sort(rules, by = "support")[1:15])
inspect(sort(rules, by = "confidence")[1:15])
inspect(sort(rules, by = "lift")[1:15])

plot(rules[1:20], method = "graph", control = list(type = "items"))
plot(rules[1:20], method = "grouped")

```

```r
plot(sort(rules, by = "support")[1:20], method = "graph", control = list(reorder = TRUE))
```

```r
plot(sort(rules, by = "confidence")[1:20], method = "graph", control = list(reorder = TRUE))
```

```r
plot(sort(rules, by = "lift")[1:20], method = "graph", control = list(reorder = TRUE))
```
